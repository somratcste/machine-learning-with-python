```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic_data = pd.read_csv("titanic.csv")
titanic_data = titanic_data.set_index("PassengerId")
```

```python
titanic_data.head(3)
```

```python
titanic_data.describe()
```

```python
titanic_data.hist(bins=20, figsize=(18,16))
```

```python
titanic_data["Sex"].value_counts()
```

```python
titanic_data["Sex"].value_counts(normalize=True)
```

```python
titanic_data["Embarked"].value_counts().plot(kind="bar", grid=True, color="#ff00ee", title="Where people embarked")
```

```python
titanic_data.corr()
```

```python
corr_matrix = titanic_data.corr()
```

```python
fig, axes = plt.subplots(figsize=(8,8))
cax = axes.matshow(corr_matrix, vmin=-1, vmax=1, cmap=plt.cm.pink)
fig.colorbar(cax)
ticks = np.arange(0, len(corr_matrix), 1)
axes.set_xticks(ticks)
axes.set_yticks(ticks)
axes.set_xticklabels(corr_matrix)
axes.set_yticklabels(corr_matrix)
plt.show()
```
